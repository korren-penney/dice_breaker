```python
# The purpose of this program is to use the random module in python 
# to roll a dice for our icebreaker today. Korren is going to attempt
# to do this live, be prepared for typos.

# import our random library
import random

# declare variables to determine the smallest and largest values we want selected
min = 1
max = 6

# declare a var to keep track of if we need to roll again
# set yes to start the program
roll_again = "yes"

# start our loop
while roll_again == "yes":
    # print a message to indicated the die is rolling
    print('Rolling the die...')
    
    # declare a var ro assign the random number
    die_value = (random.randint(min,max))
    
    # print the random number
    print("The value is: ", die_value)
    
    # create the if statement to tell the program what to ask for each number
    if die_value == 1:
        print("If you were invisible for a day what would you do?")
    elif die_value == 2:
        print("If you could choose a super power, what would you pick?")
    elif die_value == 3:
        print("What is the grossest food that you have ever eaten?")
    elif die_value == 4:
        print("Who is your favourite book or movie character?")
    elif die_value == 5:
        print("What is your favourite game to play?")
    else:
        print("What is one thing you are really good at?")
        
    # Prompt the users to see if they want to roll again
    roll_again = input("Roll again? yes/no? ")
    
print("This message will only print at the end.")
```
