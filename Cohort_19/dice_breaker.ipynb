```python
# The purpose of this program is to use the random module in python 
# to roll a dice for our icebreaker today. Korren is going to attempt
# to do this live, be prepared for typos.

# import our random library
import random

# declare our variables to determine the smallest and largest numbers
min = 1
max = 6

# declare variable to keep track of if we need to roll again
# set it to yes to start program
roll_again = "yes"

# start our loop
while roll_again == 'yes':
    # print message to indicate the die is rolling
    print("The die is rolling...")
    
    # declare var for the die value and assign it a random int
    die_value = (random.randint(min, max))
    
    # print the random number
    print("The value is: ", die_value, "You question is: ")
    
    # create our if statement to ask the questions
    if die_value == 1:
        print("If you were invisible for a day, what would you do?")
    elif die_value == 2:
        print("If you could choose a super power what would it be?")
    elif die_value == 3:
        print("What is your favourite food?")
    elif die_value == 4:
        print("Who is your favourite book or movie character?")
    elif die_value == 5:
        print("What is your favourite game to play?")
    else:
        print("What is one thing you are good at?")
        
    # Prompt the user to see if we want to roll again
    roll_again = input("Roll again? yes/no: ")
    
	# Ending the program and error handling 
    if roll_again == "no":
        print("you have selected no, program terminating")
    elif roll_again != "yes" or roll_again != "no":
        print("invalid entry, program terminating")
    
# message to show the progam has ended
print("This message will display at the end")
    
```
